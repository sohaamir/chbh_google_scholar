```python
import pandas as pd
import matplotlib.pyplot as plt

# Plotting everyone's citations per year

# Load the data from the CSV file
df = pd.read_csv('citation_statistics.csv')

# Exclude the column for citations in 2024 and sort the remaining year columns
year_columns = sorted([col for col in df.columns if col.startswith('citations_') and not col.endswith('2024')], key=lambda x: int(x.split('_')[1]))

# Create a figure and axis for the plot
plt.figure(figsize=(14, 8))

# Plot the citations per year for each row (author)
for index, row in df.iterrows():
    plt.plot([int(year.split('_')[1]) for year in year_columns], [row[year] for year in year_columns], label=row['name'])

# Configure plot settings
plt.title('Citations per Year for Each Author')
plt.xlabel('Year')
plt.ylabel('Citations')
plt.xticks(rotation=45)
plt.legend(loc='upper left', bbox_to_anchor=(1, 1))
plt.tight_layout()

# Save the plot to a file
plt.savefig('citations_plot.png')

# Show the plot
plt.show()
```

```python
import pandas as pd
from prophet import Prophet

# Load the data from the CSV file
data = pd.read_csv('citation_statistics.csv')

# Remove the 'citations_2024' column
data = data.drop(columns=['citations_2024'])

# Function to perform forecasting for each person
def forecast_citations(row):
    # Extract the relevant citation columns
    citation_cols = [col for col in row.index if col.startswith('citations_')]
    
    # Create a dataframe with the citation data
    citation_data = pd.DataFrame({'ds': citation_cols, 'y': row[citation_cols]})
    
    # Remove rows with zero citations
    citation_data = citation_data[citation_data['y'] != 0]
    
    # Convert the 'ds' column to datetime by extracting the year from the column name
    citation_data['ds'] = pd.to_datetime(citation_data['ds'].str.split('_').str[1], format='%Y')
    
    # Create a Prophet model
    model = Prophet()
    
    # Fit the model to the citation data
    model.fit(citation_data)
    
    # Generate future dates for forecasting
    future_dates = pd.DataFrame({'ds': ['2024-01-01', '2025-01-01', '2026-01-01']})
    
    # Make predictions for the future dates
    forecast = model.predict(future_dates)
    
    # Extract the forecasted values
    predicted_values = forecast['yhat'].values
    
    return pd.Series(predicted_values, index=['prediction_2024', 'prediction_2025', 'prediction_2026'])

# Apply the forecasting function to each row
data[['prediction_2024', 'prediction_2025', 'prediction_2026']] = data.apply(forecast_citations, axis=1)

# Get the citation columns
citation_cols = [col for col in data.columns if col.startswith('citations_')]

# Sort the citation columns in ascending order
citation_cols = sorted(citation_cols, key=lambda x: int(x.split('_')[1]))

# Create a new DataFrame to store the cumulative citations
cumulative_data = pd.DataFrame(index=data.index)

# Calculate the cumulative citations for each year
for i in range(len(citation_cols)):
    if i == 0:
        cumulative_data[f'cumulative_{citation_cols[i].split("_")[1]}'] = data[citation_cols[i]]
    else:
        cumulative_data[f'cumulative_{citation_cols[i].split("_")[1]}'] = cumulative_data[f'cumulative_{citation_cols[i-1].split("_")[1]}'] + data[citation_cols[i]]

# Calculate the cumulative citations for the predicted years
cumulative_data['cumulative_2024'] = cumulative_data[f'cumulative_{citation_cols[-1].split("_")[1]}'] + data['prediction_2024']
cumulative_data['cumulative_2025'] = cumulative_data['cumulative_2024'] + data['prediction_2025']
cumulative_data['cumulative_2026'] = cumulative_data['cumulative_2025'] + data['prediction_2026']

# Concatenate the original data with the cumulative citations
updated_data = pd.concat([data, cumulative_data], axis=1)

# Save the updated data to the same CSV file
updated_data.to_csv('citation_statistics_with_predictions.csv', index=False)
```

```python
import pandas as pd
import plotly.graph_objects as go
import os

# Load the data from the CSV file
data = pd.read_csv('citation_statistics_with_predictions.csv')

# Rename the prediction columns
data = data.rename(columns={'prediction_2024': 'citations_2024',
                            'prediction_2025': 'citations_2025',
                            'prediction_2026': 'citations_2026'})

# Round up all citation numbers to the closest integer
citation_cols = [col for col in data.columns if col.startswith('citations_') or col.startswith('cumulative_')]
data[citation_cols] = data[citation_cols].round().astype(int)

# Select the data for a specific person
person_name = 'Lei Zhang'
person_data = data[data['name'] == person_name].iloc[0]

# Extract the cumulative citation columns and prediction columns
cumulative_cols = [col for col in person_data.index if col.startswith('cumulative_') and person_data[col] != 0]
cumulative_prediction_cols = ['cumulative_2024', 'cumulative_2025', 'cumulative_2026']

# Create a dataframe for the cumulative citations line graph
line_data = pd.DataFrame({'Year': cumulative_cols + cumulative_prediction_cols,
                          'Citations': person_data[cumulative_cols].tolist() + person_data[cumulative_prediction_cols].tolist(),
                          'Type': ['Actual'] * len(cumulative_cols) + ['Predicted'] * len(cumulative_prediction_cols)})
line_data['Year'] = pd.to_datetime(line_data['Year'].str.split('_').str[1], format='%Y')

# Create the cumulative citations line graph with a single trace
line_fig = go.Figure(data=go.Scatter(x=line_data['Year'], y=line_data['Citations'], mode='lines+markers'))

# Add vertical line and annotation to indicate the start of predicted data
prediction_start_year = line_data[line_data['Type'] == 'Predicted']['Year'].min()
line_fig.add_shape(type='line', x0=prediction_start_year, x1=prediction_start_year, y0=0, y1=line_data['Citations'].max(),
                   line=dict(color='gray', width=1, dash='dash'))
line_fig.add_annotation(x=prediction_start_year, y=line_data['Citations'].max(), text='Prediction Start',
                        showarrow=False, yshift=10)

line_fig.update_layout(title=f'Cumulative Citations for {person_name}',
                       xaxis_title='Year',
                       yaxis_title='Cumulative Citations')

# Extract the citation columns for the bar chart
citation_cols = [col for col in person_data.index if col.startswith('citations_') and person_data[col] != 0]

# Create a dataframe for the citations by year bar chart
bar_data = pd.DataFrame({'Year': citation_cols,
                         'Citations': person_data[citation_cols].tolist()})

bar_data['Year'] = pd.to_datetime(bar_data['Year'].str.split('_').str[1], format='%Y')

# Create the citations by year bar chart
bar_fig = go.Figure(data=go.Bar(x=bar_data['Year'], y=bar_data['Citations'], text=bar_data['Citations'], textposition='auto'))

bar_fig.update_layout(title=f'Citations per Year for {person_name}',
                      xaxis_title='Year',
                      yaxis_title='Citations')

# Create the 'plots' subfolder if it doesn't exist
os.makedirs('plots', exist_ok=True)

# Save the cumulative citations line graph
line_fig.write_html(f'plots/{person_name}_cumulative_citations_1985_2026.html')

# Save the citations by year bar chart
bar_fig.write_html(f'plots/{person_name}_citations_per_year_1985_2026.html')

# Display the cumulative citations line graph in the notebook
line_fig.show()

# Display the citations by year bar chart in the notebook
bar_fig.show()
```

```python
import pandas as pd
from prophet import Prophet

# Load the data from the CSV file
data = pd.read_csv('citation_statistics_with_predictions.csv')

def forecast_citations_2023(row):
    # Extract the relevant citation columns
    citation_cols = [col for col in row.index if col.startswith('citations_') and col != 'citations_2023' and row[col] != 0]
    
    # Create a dataframe with the citation data
    citation_data = pd.DataFrame({'ds': citation_cols, 'y': row[citation_cols]})
    
    # Convert the 'ds' column to datetime by extracting the year from the column name
    citation_data['ds'] = pd.to_datetime(citation_data['ds'].str.split('_').str[1], format='%Y')
    
    # Create a Prophet model
    model = Prophet()
    
    # Fit the model to the citation data
    model.fit(citation_data)
    
    # Generate future dates for forecasting
    future_dates = pd.DataFrame({'ds': ['2023-01-01']})
    
    # Make predictions for the future dates
    forecast = model.predict(future_dates)
    
    # Extract the forecasted value
    predicted_value = forecast['yhat'].values[0]
    
    return predicted_value

# Apply the forecasting function to each row
data['predicted_citations_2023'] = data.apply(forecast_citations_2023, axis=1)

# Create the prediction_2023 DataFrame
prediction_2023 = data[['name', 'predicted_citations_2023', 'citations_2023']].copy()

# Calculate the percentage difference
prediction_2023.loc[:, 'percentage_difference'] = ((prediction_2023['citations_2023'] - prediction_2023['predicted_citations_2023']) / prediction_2023['predicted_citations_2023']) * 100

# Calculate the accuracy for each row
prediction_2023.loc[:, 'accuracy'] = 100 - abs(prediction_2023['percentage_difference'])

# Calculate the mean accuracy across all rows
mean_accuracy = prediction_2023['accuracy'].mean()
```
