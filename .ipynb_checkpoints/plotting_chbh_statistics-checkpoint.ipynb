```python
import pandas as pd
import matplotlib.pyplot as plt

# Plotting everyone's citations per year

# Load the data from the CSV file
df = pd.read_csv('citation_statistics.csv')

# Exclude the column for citations in 2024 and sort the remaining year columns
year_columns = sorted([col for col in df.columns if col.startswith('citations_') and not col.endswith('2024')], key=lambda x: int(x.split('_')[1]))

# Create a figure and axis for the plot
plt.figure(figsize=(14, 8))

# Plot the citations per year for each row (author)
for index, row in df.iterrows():
    plt.plot([int(year.split('_')[1]) for year in year_columns], [row[year] for year in year_columns], label=row['name'])

# Configure plot settings
plt.title('Citations per Year for Each Author')
plt.xlabel('Year')
plt.ylabel('Citations')
plt.xticks(rotation=45)
plt.legend(loc='upper left', bbox_to_anchor=(1, 1))
plt.tight_layout()

# Save the plot to a file
plt.savefig('citations_plot.png')

# Show the plot
plt.show()
```

```python
# Plotting those who I kinda know

# Load the DataFrame from the CSV file (the path should be set where the CSV is located)
df = pd.read_csv('citation_statistics.csv')

# Define the authors of interest
authors_of_interest = [
    "Lei Zhang",
    "Patricia Lockwood",
    "Matthew Apps",
    "Stephane De Brito",
    "Joseph Galea",
    "Katja Kornysheva",
    "Paul Muhle-Karbe",
    "Tom Rhys Marshall",
    "Romy Froemer",  # Assuming "Romy Fromer" is a typo
    "Andrew Quinn",
    "Arkady Konovalov"
]

# Filter the DataFrame to include only the authors of interest
filtered_df = df[df['name'].isin(authors_of_interest)]

# Extract the year columns, starting from 2005 to the latest available year before 2024
start_year = 2005
end_year = 2023
year_columns = [f'citations_{year}' for year in range(start_year, end_year + 1)]

# Plotting
plt.figure(figsize=(14, 8))

# Plot citations per year for each author in the filtered DataFrame
for _, row in filtered_df.iterrows():
    plt.plot(range(start_year, end_year + 1), row[year_columns], label=row['name'])
    # Annotate the last point with the author's initials
    plt.text(end_year, row[f'citations_{end_year}'], f"{row['name'].split()[0][0]}.{row['name'].split()[-1]}")

# Set the title and labels
plt.title('Citations per Year for Selected Authors (2005 onwards)')
plt.xlabel('Year')
plt.ylabel('Number of Citations')
plt.legend(loc='upper left', bbox_to_anchor=(1, 1))
plt.grid(True)
plt.tight_layout()

# Display the plot
plt.show()
```
